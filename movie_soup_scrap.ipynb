```python
# import beautiful soup

from bs4 import BeautifulSoup as bs
import requests
```

```python
#url page to be scrape

url = "https://en.wikipedia.org/wiki/List_of_Academy_Award-winning_films"
```

```python
# request page with requests modules
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
# results are returned as an iterable list

title_results = soup.find_all('a')
```

```python
# check if the title results hold wanted elements
print(title_results)
```

```python
# loop through the data to extract title
for title in title_results:
    title = title_results.find_all('a', class_='title')
    print(title)
    
```

# use pandas table scraping method

```python
import pandas as pd
```

```python
tables = pd.read_html(url)
tables
```

```python
# pandas picked up some useless forms 
df = tables[0]
```

```python
# show the table
df.head()
```

```python
#changed the columns title
df.columns = ['Film','Year','Awards','Nominations']
df.head()
```

```python
# deleted the first line
df = df.iloc[1:]
df.head()
```

```python
# store
df_awards = df.sort_values("Awards", ascending = False)
```
