```python
# Dependencies
import requests
import json
import pandas as pd
from pprint import pprint
from bs4 import BeautifulSoup
import os
```

```python
# Note that the ?t= is a query param for the t-itle of the
# movie we want to search for.
url = "http://www.omdbapi.com/?t="
api_key = "&apikey=trilogy"
```

```python
# Collect Oscars data from CSV File
csv_file = "database.csv"
movies = pd.read_csv(csv_file)
movies_df = pd.DataFrame(movies)
movies_df.head()
```

```python
movies_df = movies_df.dropna()
movies_df
```

```python
unique_movies = (list(movies_df['Film'].unique()))
unique_movies
```

```python
movies = []
movie = movies_df['Film']
for x in range(len(unique_movies)):
    if movie not in movies:
        movies.append(movie)
movies
```

```python
response_json = []
# Make a request for each of the indices
for x in range(len(unique_movies)):
    unique_movies[x] = unique_movies[x].replace(" ", "_")
    full_url = f"{url + unique_movies[x]}&appid={api_key}" 
    print(full_url)
    # Get one of the posts
    post_response = requests.get(full_url)
    # Save post's JSON
    response_json.append(post_response.json())
    print(f"Processing Record {x} of Set 1 | {unique_movies[x]}.")
```

```python
pprint(response_json)
```

```python
response_json_df = pd.DataFrame(response_json)
response_json_df
```

```python
# Converting the response to JSON, and printing the result.
data = response_json
pprint(data)
```
