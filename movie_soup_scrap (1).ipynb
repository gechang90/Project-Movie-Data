```python
# import beautiful soup

from bs4 import BeautifulSoup as bs
import requests
# Dependencies
import json
import pandas as pd
from pprint import pprint
import os
from sqlalchemy import create_engine
```

```python
#url page to be scrape

url = "https://en.wikipedia.org/wiki/List_of_Academy_Award-winning_films"
```

```python
# request page with requests modules
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
# results are returned as an iterable list

title_results = soup.find_all('a')
```

```python
# check if the title results hold wanted elements
print(title_results)
```

```python
tables = pd.read_html(url)
tables
```

```python
# show the table
df.head()
```

```python
#changed the columns title
df.columns = ['Film','Year','Awards','Nominations']
df.head()
```

```python
# store
df_awards = df.sort_values("Awards", ascending = False)
```

```python
df_awards.to_csv("df_awards2.csv",index = False)
```

```python
# Note that the ?t= is a query param for the t-itle of the
# movie we want to search for.
url2 = "http://www.omdbapi.com/?t="
api_key = "&apikey=trilogy"
```

```python
#df_awards['Film_url']=df_awards['Film_url'].replace({" ":"_"})
#df_awards
```

```python
unique_movies = list((df_awards['Film'].unique()))
unique_movies
```

```python
movies = []
movie = list(df_awards['Film'])
for x in range(len(df_awards['Film'])): 
    if movie not in movies:
        movies.append(movie)
movies
test = movies[0]
test
```

```python
#test = test.replace(' ',"_")
movies
newtest = []
for x in test:
    x = x.replace(' ','_')
    newtest.append(x)
newtest
#movies_df = pd.DataFrame(movies)
#movies_df.str.replace({" ":"_"}).unstack()
#movies_df
```

```python
response_json = []
# Make a request for each of the indices
for x in range(len(newtest)):
    full_url = f"{url2 + newtest[x]}&appid={api_key}"
    print(full_url)
    # Get one of the posts
    post_response = requests.get(full_url)
    # Save post's JSON
    response_json.append(post_response.json())
    print(f"Processing Record {x} of Set 1 | {newtest[x]}.")
```

```python
pprint(response_json)
```

```python
OMDB_response_df = pd.DataFrame(response_json)
OMDB_response_df
```

```python
OMDBAPI_df = OMDB_response_df[['Title', 'Metascore', 'imdbRating']]
OMDBAPI_df
```
